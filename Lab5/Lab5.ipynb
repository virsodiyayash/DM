### 2)  Handle Missing Values in data set [use dropna(), fillna(), and interpolate]


```python
df.dropna(how="any" , axis=0)
```

```python
newData = df.copy()
newData["Age"] = df["Age"].fillna(value=df["Age"].mean())
newData.isnull().sum()

newData.info()


```

```python
newData.fillna({"Cabin" : "Not Available"} , inplace=True)
newData["Cabin"]
```

### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
pureData = df.copy()
```

```python
pure_data_2 = df.copy()
age_nm = pure_data_2["Age"]
age_min = age_nm.min()
age_max = age_nm.max()

pure_data_2["MinMaxAge"] = (age_nm - age_max)/(age_max - age_min)
pure_data_2["MinMaxAge"]
```

```python
hello = df.copy()
hello["Age"] = hello["Age"].interpolate()
hello["New Age"] = hello["Age"] / 10 ** len(str(int(hello["Age"].max())))
hello[["Age" , "New Age"]]
```
